# 3. 데이터 전처리

## 3.1 결측치 처리하기

```python
# Seaborn 라이브러리 임포트
import seaborn as sns

# Pandas 라이브러리를 사용하여 CSV 파일 불러오기
titanic_data = sns.load_dataset('titanic')
titanic_data.head()  # 데이터의 처음 5행 출력
```

```python
# 결측치 확인
missing_values = titanic_data.isnull().sum()
missing_values
```

```python
# 'age' 컬럼의 결측치를 중앙값으로 대체
titanic_data['age'].fillna(titanic_data['age'].median(), inplace=True)

# 결측치 재확인
titanic_data.isnull().sum()
```

## 3.2 데이터 타입 변환하기

```python
# 'sex' 컬럼의 데이터 타입을 문자열에서 숫자로 변환 (male: 0, female: 1)
titanic_data['sex'] = titanic_data['sex'].map({'male': 0, 'female': 1})
titanic_data.head()
```

## 3.3 필요한 열만 선택하기

```python
# 'survived', 'pclass', 'sex', 'age' 열만 선택
selected_data = titanic_data[['survived', 'pclass', 'sex', 'age']]
selected_data.head()
```
